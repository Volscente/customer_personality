```python
# Import Standard Modules
import pandas as pd

import plotly.express as ex

# Set Pandas Options
pd.set_option('display.max_columns', 500)
```

# Read Data

```python
# Read data from csv
data = pd.read_csv('../data/marketing_campaign.csv', sep='\t', encoding='latin1')
```

# Feature Exploration

## Year Birth

```python
# Check for NaN values
print('Year_Birth NaN value count: {}'.format(len(data[data['Year_Birth'].isna()])))
```

```python
# Plot the Histogram of Year_Birth Distribution
figure = ex.histogram(data, 
                      x='Year_Birth', 
                      title='Year Birth Distribution', 
                      labels={'x': 'Year Birth', 'y': 'Count'}, 
                      template='plotly_dark')

figure.show()
```

```python
# Plot the Boxplot of Year_Birth Distribution
figure = ex.box(data_frame=data, 
                y='Year_Birth', 
                template='plotly_dark')

figure.show()
```

There are 3 outliers with respect to the 'Year_Birth'.
